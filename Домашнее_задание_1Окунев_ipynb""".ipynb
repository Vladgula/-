2. **Столбчатая диаграмма для категориальной переменной `Class`:**

```python
# Определяем список категориальных колонок (как мы определили ранее)
# Разделим их на порядковые и номинальные для примера, хотя countplot работает для всех
ordinal_cols = [
    'Parental_Involvement', 'Access_to_Resources', 'Motivation_Level',
    'Family_Income', 'Teacher_Quality', 'Peer_Influence',
    'Parental_Education_Level', 'Distance_from_Home'
]
nominal_cols = [
    'Extracurricular_Activities', 'Internet_Access', 'School_Type',
    'Learning_Disabilities', 'Gender'
]

categorical_cols = ordinal_cols + nominal_cols # Объединяем все категориальные

# Устанавливаем стиль графиков
sns.set_theme(style="whitegrid")

print("--- Столбчатые диаграммы для категориальных переменных ---")

# Перебираем каждую категориальную колонку и строим countplot
for col in categorical_cols:
    if col in df.columns:
        plt.figure(figsize=(10, 6)) # Размер графика

        # Строим countplot
        # palette='viridis' - цветовая схема, можно менять
        sns.countplot(data=df, x=col, palette='viridis', order=df[col].value_counts().index)
        # order=df[col].value_counts().index - сортирует столбцы по убыванию частоты, что часто удобнее

        plt.title(f'Распределение переменной "{col}"')
        plt.xlabel(col)
        plt.ylabel('Количество наблюдений (частота)')
        plt.xticks(rotation=45, ha='right') # Поворачиваем метки на оси X, если они длинные
        plt.tight_layout() # Автоматически подгоняет параметры графика, чтобы избежать наложений
        plt.show() # Отображаем график
    else:
        print(f"\nПредупреждение: Колонка '{col}' не найдена в DataFrame.")
```
